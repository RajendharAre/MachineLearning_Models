### Kmeans Clustering

#### importing the libraries



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### importing the dataset

```python
dataset = pd.read_csv('../Datasets/Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

#### Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
# plt.label()
plt.show()
```

#### Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
```

#### Visualising the clusters

```python
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
plt.title('Clusters of Customers')
plt.legend()
plt.show()
```

#### Final Conclusion

1. **Cluster 1: Moderate Income, Moderate Spending**
    - Customers with average annual income and average spending scores.
    - Likely to be steady, regular shoppers.
    - *Improvement Suggestion:* Offer loyalty programs, personalized discounts, and exclusive events to encourage more frequent visits and higher spending.

2. **Cluster 2: High Income, Low Spending**
    - Customers with high annual income but low spending scores.
    - Potentially value-conscious or less engaged with mall offerings.
    - *Improvement Suggestion:* Introduce premium services, luxury brand collaborations, and targeted marketing to showcase exclusive products and experiences.

3. **Cluster 3: Low Income, High Spending**
    - Customers with lower annual income but high spending scores.
    - May represent enthusiastic shoppers or those prioritizing mall purchases.
    - *Improvement Suggestion:* Provide budget-friendly deals, installment payment options, and reward schemes to maintain their enthusiasm and spending levels.

4. **Cluster 4: Low Income, Low Spending**
    - Customers with low annual income and low spending scores.
    - Less likely to make frequent or high-value purchases.
    - *Improvement Suggestion:* Organize community events, offer affordable product bundles, and provide incentives for referrals to increase engagement and spending.

5. **Cluster 5: High Income, High Spending**
    - Customers with high annual income and high spending scores.
    - Most valuable segment, likely to be premium shoppers.
    - *Improvement Suggestion:* Deliver personalized VIP experiences, early access to new collections, and exclusive privileges to retain loyalty and further boost spending.
